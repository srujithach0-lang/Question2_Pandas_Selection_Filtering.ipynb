Pandas Employee Data Analysis
# 1. Create a pandas DataFrame using the employee **data**

```python
import pandas as pd

data = {
    'EmployeeID': [201, 202, 203, 204, 205, 206, 207, 208, 209, 210],
    'Name': ['Sarah', 'Michael', 'Jennifer', 'David', 'Lisa', 'James', 'Emma', 'Robert', 'Sophia', 'William'],
    'Department': ['Engineering', 'Sales', 'Marketing', 'Engineering', 'HR', 'Sales', 'Engineering', 'Marketing', 'Sales', 'HR'],
    'Salary': [85000, 62000, 71000, 92000, 58000, 68000, 78000, 75000, 65000, 60000],
    'Performance Score': [4, 3, 5, 5, 4, 4, 3, 4, 5, 3],
    'Years of Service': [5, 3, 7, 8, 4, 6, 2, 5, 4, 3]
}

df = pd.DataFrame(data)
df.set_index('EmployeeID', inplace=False) # Optional: keeps EmployeeID as a column or index
df
```

# 2.Select and display only the 'Name', 'Department', and 'Salary' **columns**

```python
df[['Name', 'Department', 'Salary']]
```

# **3. Use iloc to display the employee at row position 5**

```python
# Note: iloc is zero-based, so position 5 is the 6th row
df.iloc[5]
```

# 4. Use loc to display the employee with EmployeeID **208**

```python
# We filter the row where EmployeeID matches 208
df.loc[df['EmployeeID'] == 208]
```

# 5. Filter and display all employees from the 'Engineering' **department**

```python
df[df['Department'] == 'Engineering']
```

# **6. Filter and display all employees with Salary greater than 70000**

```python
df[df['Salary'] > 70000]
```

# 7. Filter and display employees from 'Sales' department with Performance Score >= **4**

```python
df[(df['Department'] == 'Sales') & (df['Performance Score'] >= 4)]
```

# **8. Sort the entire DataFrame by Salary in descending order and display the top 3 highest-paid employees**

```python
df.sort_values(by='Salary', ascending=False).head(3)
```

# **9. Sort by Department (ascending) and then by Salary (descending) within each department**

```python
df.sort_values(by=['Department', 'Salary'], ascending=[True, False])
```
